### --> number of items fold by Item Fat Content wise

```python
print("Item Fat Content wise : ",)
Item_Fat_Content_wise=df.groupby(['Item Fat Content']).agg(No_of_items_sales =("Sales","count"),Total_sales=("Sales",'sum')).reset_index()
Item_Fat_Content_wise
```

### --> number of items sold byOutlet Type wise

```python
print("Outlet Type wise : ")
Outlet_Type_wise=df.groupby(['Outlet Type']).agg(Total_items = ("Sales","count"),Total_sales=("Sales",'sum')).reset_index()
Outlet_Type_wise
```

### -->Year wise sales

```python
year=df.groupby(['Outlet Establishment Year']).agg(Year_wise_sales=("Sales",'sum')).reset_index()
year
```

## Item wise total_sales

```python
Item_Type=df.groupby(['Item Type']).agg(Total_Sale=("Sales",'sum')).reset_index().sort_values("Total_Sale",ascending= False )
Item_Type
```

# Charts Requrements

```python
items=df.groupby(['Item Fat Content']).agg(Total_sales=("Sales",'sum')).reset_index()
items
```

## -->Sales Distribution by Item Fat Content

```python
# Assuming 'items' is your DataFrame
labels = items['Item Fat Content']
sizes = items['Total_sales']

# Create pie chart
plt.figure(figsize=(4,6))
plt.pie(sizes, labels=labels, 
        autopct='%.1f%%', 
        startangle=90)
plt.title("Sales Distribution by Item Fat Content")
plt.show()
```

## -->sales Distribution by Outlet Type

```python
labels = Outlet_Type_wise['Outlet Type']
values= Outlet_Type_wise['Total_sales']

plt.figure(figsize=(8,4))
plt.bar(labels,
       values,
       color=['skyblue', 'lightgreen']
       )
plt.title("sales Distribution by Outlet Type ")
plt.xlabel('Outlet Type')
plt.ylabel('Total Sales')
plt.show()



```

## -->year wise sales

```python
x=year['Outlet Establishment Year']
y=year['Year_wise_sales']
lables = ('year wise sales')

plt.plot(x,y,marker='o',color='r',linestyle=':',label =['year wise sales'])


plt.xlabel("Outlet Establishment Year")
plt.ylabel("Year_wise_sales")
plt.title('year wise sales')
plt.legend()

plt.show
```
